# Multi-Feature Regression
ML-python/2-regression-multi-variable

## Tutorial

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

### Missing data assumption
Using Median

```python
df.bedrooms.median()
print(df.bedrooms.median())
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

### Model Fitting

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

```python
print((reg.coef_,reg.intercept_))
print(reg.predict([[3000, 3, 40]]))
112.06244194*3000 + 23388.88007794*3 + -3231.71790863*40 + 221323.00186540384
```

```python
import matplotlib.pyplot as plt
xaxis = 'bedrooms'
%matplotlib inline
plt.xlabel(xaxis)
plt.ylabel('price')
plt.scatter(df[[xaxis]],df.price,color='red',marker='+')
plt.plot(df[[xaxis]],reg.predict(df[['area','bedrooms','age']]))
```

```python
xaxis = 'age'
%matplotlib inline
plt.xlabel(xaxis)
plt.ylabel('price')
plt.scatter(df[[xaxis]],df.price,color='red',marker='+')
plt.plot(df[[xaxis]],reg.predict(df[['area','bedrooms','age']]))
```
