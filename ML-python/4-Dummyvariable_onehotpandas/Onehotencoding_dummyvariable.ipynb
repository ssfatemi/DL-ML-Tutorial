# One-Hot encoding

## Solution 1
Using Panda's features

```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
final = merged.drop(['town'], axis='columns')
final
```

```python
final = final.drop(['west windsor'], axis='columns')
final
```

```python
y = final.price
X = final.drop('price', axis='columns')
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
print(model.predict(X))

print('accuracy:')
print(model.score(X,y))
print('')
print(model.predict([[3400,0,0]])) # 3400 sqr ft home in west windsor)
```

## Solution 2
Using sklearn OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')

X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

## Exersice

```python
import pandas as pd
df = pd.read_csv("carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
final = merged.drop(['Car Model'], axis='columns')
final
```

```python
final = final.drop(['Mercedez Benz C class'], axis='columns')
final
```

```python
y = final['Sell Price($)']
X = final.drop('Sell Price($)', axis='columns')
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
# predicting the Mercedez Benz price with 4 years old and 45000 milage
print(model.predict([[45000,4,0,0]]))
```
